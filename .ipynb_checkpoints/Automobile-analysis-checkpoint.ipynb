## Data Description

The dataset used for this visualization assignment contains automobile_sales data representing automobile sales and related variables during recession and non-recession period.

The dataset includes the following variables:
1. Date: The month end date of sales observation.
2. Recession: A binary variable indicating recession perion; 1 means it was recession, 0 means it was normal.
3. Automobile_Sales: The number of vehicles sold during the period.
4. GDP: The per capita GDP value in USD.
5. Unemployment_Rate: The monthly unemployment rate.
6. Consumer_Confidence: A synthetic index representing consumer confidence, which can impact consumer spending and automobile purchases.
7. Seasonality_Weight: The weight representing the seasonality effect on automobile sales during the period. This variable represents the seasonal effect on automobile sales for a given month. In the automobile industry, sales often fluctuate throughout the year due to seasonal patterns—for example, sales may increase during the festive season or year-end promotions, and decrease during off-peak months like post-holiday winter periods. A value greater than 1 indicates higher-than-average sales expected for that month due to seasonal trends (e.g., holiday season or new model launches). A value less than 1 suggests lower-than-average sales due to seasonal slowdowns. A value around 1 means the season has neutral or average effect on sales.
8. Price: The average vehicle price during the period.
9. Advertising_Expenditure: The advertising expenditure of the company.
10.Vehicle_Type: The type of vehicles sold; Supperminicar, Smallfamiliycar, Mediumfamilycar, Executivecar, Sports.
11.Competition: The measure of competition in the market, such as the number of competitors or market share of major manufacturers.
12.Month: Month of the observation extracted from Date..
13.Year: Year of the observation extracted from Date.
By examining various factors mentioned above from the dataset, you aim to gain insights into how recessions impacted automobile sales for your company.

## 3.  Creating Visualizations for Data Analysis

Since we will be creating trends we are going to sort the data by the date so that time flows accurately and recesion periods align properly

```python
df= df.sort_values(by='Date')
```

The next option will be to reset the index now 

```python
df=df.reset_index(drop=True)
```

```python
df.head()
```

### 3.1 Line plot to show how Average automobile sales fluctuate from year to year

```python
df_line= df.groupby('Year')['Automobile_Sales'].mean()

plt.figure(figsize=(10, 6))
df_line.plot(kind= 'line')
plt.xticks(list(range(1980, 2024)), rotation=75)

#labels
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.title('Automobile Sales during Recession')

#Annotations for recession years
plt.text(1982, df_line.loc[1982], '1981-82 Recession')
plt.text(2009, df_line.loc[2009], '2008-09 Recession')

plt.grid(True)
plt.legend(['Average Sales'])
plt.show()
```

#### Insight 
This line chart illustrates the yearly average automobile sales from 1980 to 2023. The visualization highlights fluctuations in sales over time, with noticeable declines during recession periods. Annotations for the 1981–1982 and 2008–2009 recessions show clear drops in automobile sales, reflecting reduced consumer spending during economic downturns. This chart demonstrates the strong relationship between macroeconomic conditions and automobile sales performance.

```python
df_rec = df[df['Recession'] == 1]

# Aggregate sales by Year and Vehicle_Type
trend_data = df_rec.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()

# Set figure size
plt.figure(figsize=(14, 6))

# Plot lines for each vehicle type
sns.lineplot(
    data=trend_data,
    x='Year',
    y='Automobile_Sales',
    hue='Vehicle_Type',
    marker='o'
)

# Customize labels and title
plt.title("Sales Trends by Vehicle Type During Recession Periods")
plt.xlabel("Year")
plt.ylabel("Average Vehicle Sales")
plt.legend(title='Vehicle Type')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 3.2 How do trends in advertising expenditure correlate with automobile sales during non-recession periods

```python
# filter the years with recession
non_recession_df= df[df['Recession']== 0]
```

group by year and calculate averages

```python
df_trends=non_recession_df.groupby('Year', as_index=False).agg(
    Avg_Sales=('Automobile_Sales', 'mean'),
    Avg_Ad_Spend=('Advertising_Expenditure', 'mean')
)
#Create line plot using seaborn
sns.lineplot(data=df_trends,
            x='Year',
            y='Avg_Sales',
            marker='o',
            linestyle='-',
            color='green',
            label='Average Sales'
)

#Line for Average Advertising Expenditure
sns.lineplot(
    data=df_trends,
    x='Year',
    y='Avg_Ad_Spend',
    markers='s',
    linestyle='--',
    color='blue',
    label='Average Advertising Expenditure'
)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Average Automobile Sales and Advertising Expenditure during Non-Recession Periods')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

#### Insight
The line chart illustrates the trends in average automobile sales and advertising expenditure during non-recession periods. Overall, a positive relationship is observed, where higher advertising expenditure is generally associated with increased automobile sales. However, automobile sales exhibit greater volatility than advertising expenditure, indicating that sales are influenced by additional factors beyond marketing alone. The analysis suggests that advertising is most effective during stable economic conditions, serving as a supporting factor that amplifies demand rather than acting as the sole driver of sales.

## 3.3 Compare the sales trend per vehicle type for a recession period with a non recession period

```python
new_df=df.groupby('Recession')['Automobile_Sales'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.xticks(ticks=[0,1], labels=['Non-Recession', 'Recession'])
sns.barplot(x='Recession', y='Automobile_Sales', hue='Recession', data= new_df)

plt.xlabel('Recession')
plt.ylabel('Automobile Sales')
plt.title('Average Automobile Sales during Recession and Non recession period')
plt.show()
```

#### Now lets compare the sales of a different vehicle types during recession and non recession period

```python
grouped_df=(df.groupby(['Recession', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index())

plt.figure(figsize=(12, 8))
sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data= grouped_df)

plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.xlabel('Period')
plt.ylabel('Average Automobile Sales')
plt.title('Vehicle-Wise Sales during Recession and Non-Recession Periods')
plt.legend(title='Vehicle_Type')
plt.grid(axis='y', linestyle='--', alpha= 0.7)
plt.show()
```

#### Insight
From the above plot its clear that there is a decline in total sales of the automobiles during recession with the most affected vehicles being axecutive car and sports which experienced a huge decline.

### 3.4 Lets compare the variations in GDP during recession and non recession period by developing line plots for each period

```python
rec_df=df[df['Recession']==1]
non_rec_df=df[df['Recession']==0]
```

```python
fig=plt.figure(figsize=[12,6])
ax0=fig.add_subplot(1,2,1)
ax1=fig.add_subplot(1,2,2)

sns.lineplot(x='Year', y='GDP', data=rec_df, label='Recession', ax=ax0, color='red')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.set_title('GDP Variation during the Recession Period')

sns.lineplot(x='Year', y='GDP', data=non_rec_df, label='Non-Recession', ax=ax1, color='green')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.set_title('GDP Variation during the Non-Recession period')

plt.tight_layout()
plt.show()
```

#### Insight 
From this plot, it is evident that GDP tends to be lower and more volatile during recessions, with significant fluctuations and uncertainty. In contrast, non-recessionary periods show a relatively higher GDP level but still exhibit some degree of variability.


### 3.5 Lets plot a buble plot to display the impact of seasonality on automobile sales

```python
non_rec_df=df[df['Recession']==0]

size=non_rec_df['Seasonality_Weight']
plt.figure(figsize=(12, 6))

sns.scatterplot(data=non_rec_df,
               y='Automobile_Sales',
               x='Month',
               size='Seasonality_Weight',
               hue='Seasonality_Weight',
               alpha=0.7,
               palette='viridis'
               )

plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.title('Impact of seasonality on Automobile Sales')
plt.legend(title='Seasonality Weight', bbox_to_anchor=(1.05, 1), loc=('upper left'))
plt.tight_layout()
plt.show()
```

#### Insight 
From this plot, it is evident that seasonality has not affected on the overall sales. However, there is a drastic raise in sales in the month of April and December

### 3.6  Lets develop a scatter plot to identify the relation ship between consumer confidence and automobile sales during Recessions.

```python
rec_df=df[df['Recession']==1]
plt.figure(figsize=(12,6))

plt.scatter(rec_df['Consumer_Confidence'], rec_df['Automobile_Sales'])

plt.xlabel('Consumer Confidence')
plt.ylabel('Automobile Sales')
plt.title('Consumer Confidence and Automobile Sales during Recession Period')

plt.tight_layout()
plt.show()
```

### How does the consumer confidence relate to automobile sales during recessions

```python
plt.figure(figsize=(12,6))

plt.scatter(rec_df['Price'], rec_df['Automobile_Sales'])

plt.xlabel('Vehicle Price')
plt.ylabel('Automobile Sales')
plt.title('Relationship between Vehicle Price and Sales during Recessions')

plt.tight_layout()
plt.show()
```

#### Insight 
This scatter plot shows the relationship between vehicle prices and automobile sales during recession periods. The plot indicates that higher vehicle prices are generally associated with lower sales, suggesting that consumers become more price-sensitive during economic downturns. Lower-priced vehicles tend to maintain relatively higher sales volumes during recessions.

### 3.7 Lets create a pie chart to display the portion of advertising expenditure of XYZ Automotives during the recession and non-recession periods

```python
rec_df=df[df['Recession']==1]
non_rec_df=df[df['Recession']==0]

rec_total=rec_df['Advertising_Expenditure'].sum()
non_rec_total=non_rec_df['Advertising_Expenditure'].sum()

plt.figure(figsize=(9,6))

labels=['Recession', 'Non-Recession']
sizes=[rec_total, non_rec_total]
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Advertising Expenditure during Recession and Non-Recession Periods')
plt.show()
```

#### Insight
It seems XYZAutomotives has been spending much more on the advertisements during non-recession periods as compared to during recession times.

### 3.8 A pie chart to display the total Advertisement expenditure for each vehicle type during recession period

```python
rec_df=df[df['Recession']==1]

vt_exp=rec_df.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

plt.figure(figsize=(8,6))
labels=vt_exp.index
sizes=vt_exp.values
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)

plt.title('Share of Each vehicle Type in Total Expenditure during Recessions')
plt.show()
```

#### Insight 
The pie chart illustrates how advertising expenditure was distributed across different vehicle types during recession periods. It is evident that a larger share of the advertising budget was allocated to mass-market vehicle categories such as small and medium family cars, while luxury and sports vehicles received comparatively lower advertising investment. This suggests that during economic downturns, the company prioritized promoting more affordable and practical vehicles that are likely to remain in demand when consumer spending power is reduced.

### 3.9 Develop a line plot to analyze the effect of the Unemployment rate on the vehicle type and sales during the recession periods

```python
line_data = rec_df.groupby(['Date', 'Vehicle_Type']).agg({
    'Automobile_Sales': 'sum',
    'unemployment_rate': 'mean'
}).reset_index()
```

```python
# Initialize the figure
plt.figure(figsize=(12,6))

# Line plot for Sales by Vehicle Type
sns.lineplot(data=line_data, x='Date', y='Automobile_Sales', hue='Vehicle_Type', marker='o')

# Optional: Add unemployment rate as a secondary axis
ax = plt.gca()
ax2 = ax.twinx()
sns.lineplot(data=line_data, x='Date', y='unemployment_rate', color='red', ax=ax2, label='Unemployment Rate', marker='x')
ax2.set_ylabel('unemployment rate (%)')

# Labels and title
plt.title('Effect of Unemployment Rate on Vehicle Sales During Recession')
ax.set_ylabel('Automobile_Sales')
ax.set_xlabel('Date')
plt.legend(loc='upper left')

plt.show()
```

```python
# Assuming 'df' is your dataset
df_rec = df[df['Recession'] == 1]

# Set figure size
plt.figure(figsize=(16, 6))

# Create line plot showing effect of unemployment rate on automobile sales, by vehicle type
sns.lineplot(
    data=df_rec,
    x='unemployment_rate',   # unemployment rate column
    y='Automobile_Sales',                # automobile sales column
    hue='Vehicle_Type',       # vehicle type column
    marker='o'
)

# Customize labels, legend, and title
plt.title('Effect of Unemployment Rate on Vehicle Sales During Recession')
plt.xlabel('Unemployment Rate (%)')
plt.ylabel('Vehicle Sales')
plt.legend(title='Vehicle Type')

# Show plot
plt.tight_layout()
plt.show()
```

#### Insight

The plot shows that automobile sales decline as unemployment rates rise during a recession, with sharp drops beyond 3%. SuperMiniCars, SmallFamilyCars, and MediumFamilyCars exhibit high volatility, reflecting their sensitivity to economic uncertainty and changing consumer priorities.
